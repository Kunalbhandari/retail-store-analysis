```python
# creating dic of customers
dic=dict()
ctr=0
chk=df['CustomerID'].isnull()
for i in range(len(df)):
    if chk[i]==False:
        ctr+=1
        if df['CustomerID'][i] not in dic:
            dic[df['CustomerID'][i]]=dict()
            dic[df['CustomerID'][i]]['Country']=df['Country'][i]
            dic[df['CustomerID'][i]]['InvoiceNo']=set([(df['InvoiceNo'][i],df['InvoiceDate'][i])])
        else:
            dic[df['CustomerID'][i]]['InvoiceNo'].add((df['InvoiceNo'][i],df['InvoiceDate'][i]))

# print(dic)
print(ctr)

# df['InvoiceNo'].isnull().values.any()
```
